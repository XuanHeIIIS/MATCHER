# MATCHER Demo

**Import Packages**

```python
%matplotlib inline
%config InlineBackend.figure_format = 'png'
import matplotlib
import matplotlib.pyplot as plt
import GPy
import os
import numpy as np
from matcher import MATCHER
```

**Read synthetic data**

```python
lin_traj = np.genfromtxt("data/linear_trajectory.csv",delimiter=',')
sqrt_traj = np.genfromtxt("data/sqrt_trajectory.csv",delimiter=',')
sq_traj = np.genfromtxt("data/sq_trajectory.csv",delimiter=',')
logit_traj = np.genfromtxt("data/logit_trajectory.csv",delimiter=',')
master_time = np.genfromtxt("data/master_time.csv",delimiter=',')
```

**Perform manifold alignment on simulated data**

```python
X = [lin_traj,sqrt_traj,sq_traj,logit_traj]
m = MATCHER(X)
m.infer()
```

**Plot warping functions for synthetic data**

```python
ax = m.warp_functions[0].plot()
l = ax.legend()
l.remove()
plt.title(r"$t = t_m$",fontsize=20)
ax = m.warp_functions[1].plot()
l = ax.legend()
l.remove()
plt.title(r"$t = \sqrt{t_m}$",fontsize=20)
ax = m.warp_functions[2].plot()
l = ax.legend()
l.remove()
plt.title(r"$t=t_m^2$",fontsize=20)
ax = m.warp_functions[3].plot()
l = ax.legend()
l.remove()
plt.title(r"$t=1+\frac{1}{6} log \frac{t_m}{1-t_m}$",fontsize=10)
```

**Plot true vs. inferred master time**

```python
fig,axes = plt.subplots(2,2)
axes[0,0].scatter(m.master_time[0],master_time[:,0])
axes[0,1].scatter(m.master_time[1],master_time[:,1])
axes[1,0].scatter(m.master_time[2],master_time[:,2])
axes[1,1].scatter(m.master_time[3],master_time[:,3])
```
