# MATCHER Demo

**Import packages**

```python
%matplotlib inline
%config InlineBackend.figure_format = 'png'
import matplotlib
import matplotlib.pyplot as plt
import GPy
import os
import numpy as np
from matcher import MATCHER
```

**Read input data**

```python
os.chdir('/proj/prins_lab/Josh/CellIntegration/')
rna_data = np.genfromtxt("mESC_scB&T_RNA.csv",delimiter=',')
rna_data = rna_data.transpose()
rna_data2 = np.genfromtxt("mESC_RNA_250_cells_169_genes_6_signatures_scaled.csv",delimiter=',')
#rna_data2 = np.genfromtxt("mESC_RNA_250_cells_180_genes.csv",delimiter=',')
#rna_data2 = rna_data2.transpose()
rna_data3 = np.genfromtxt("mESC_RNA_183_cells_6661_genes.csv",delimiter=',')

met_data = np.genfromtxt("methylation_data.csv",delimiter=',')
met_data = met_data.transpose()
#met_data = np.genfromtxt("mESC_DNA_met_61_cells_2872_sites.csv",delimiter=',')
#atac_data = np.genfromtxt("mESC_GC_norm_TF_counts_77_cells.csv",delimiter=',')
atac_data = np.genfromtxt("mESC_norm_TF_counts_76_cells_192_features.csv",delimiter=',')
#atac_data3 = np.genfromtxt("mESC_normalized_deviation.csv",delimiter=',')
chip_data = np.genfromtxt("mESC_H3K4me2_4587_cells_91_signatures_norm.csv",delimiter=',')
#chip_data = np.genfromtxt("mESC_H3K4me2_4587_cells_186_TFs_1_signature_norm.csv",delimiter=',')
lin_traj = np.genfromtxt("linear_trajectory.csv",delimiter=',')
sqrt_traj = np.genfromtxt("sqrt_trajectory.csv",delimiter=',')
sq_traj = np.genfromtxt("sq_trajectory.csv",delimiter=',')
logit_traj = np.genfromtxt("logit_trajectory.csv",delimiter=',')
master_time = np.genfromtxt("master_time.csv",delimiter=',')
t0 = np.genfromtxt("noise_trajectory_0.csv",delimiter=',')
t1 = np.genfromtxt("noise_trajectory_1.csv",delimiter=',')
t3 = np.genfromtxt("noise_trajectory_3.csv",delimiter=',')
t5 = np.genfromtxt("noise_trajectory_5.csv",delimiter=',')
t7 = np.genfromtxt("noise_trajectory_7.csv",delimiter=',')
t9 = np.genfromtxt("noise_trajectory_9.csv",delimiter=',')
```

**Perform manifold alignment on simulated data**

```python
X = [lin_traj,sqrt_traj,sq_traj,logit_traj,t0,t1,t3,t5,t7,t9]
m = matcher.MATCHER(X)
m.infer()
```

**Plot warping functions for synthetic data**

```python
ax = m.warp_functions[0].plot()
l = ax.legend()
l.remove()
plt.title(r"$t = t_m$",fontsize=20)
plt.savefig("lin_traj_warp.svg",format="svg",dpi=1000)
ax = m.warp_functions[1].plot()
l = ax.legend()
l.remove()
plt.title(r"$t = \sqrt{t_m}$",fontsize=20)
plt.savefig("sqrt_traj_warp.svg",format="svg",dpi=1000)
ax = m.warp_functions[2].plot()
l = ax.legend()
l.remove()
plt.title(r"$t=t_m^2$",fontsize=20)
plt.savefig("sq_traj_warp.svg",format="svg",dpi=1000)
ax = m.warp_functions[3].plot()
l = ax.legend()
l.remove()
plt.title(r"$t=1+\frac{1}{6} log \frac{t_m}{1-t_m}$",fontsize=10)
plt.savefig("logit_traj_warp.svg",format="svg",dpi=1000)
```

```python
m.warp_functions[4].plot()
m.warp_functions[5].plot()
m.learn_warp_function(6,reverse=True)
m.warp_functions[6].plot()
m.learn_warp_function(7,reverse=True)
m.warp_functions[7].plot()
m.learn_warp_function(8,reverse=True)
m.warp_functions[8].plot()
m.learn_warp_function(9,reverse=True)
m.warp_functions[9].plot()
```

**Perform manifold alignment on real data**

```python
X = [rna_data2,met_data,atac_data/1000.,chip_data*10.]
m = matcher.MATCHER(X)
m.infer()
m.learn_warp_function(1,reverse=True)
m.learn_warp_function(2,reverse=True)
```

**Plot warping functions for real data**

```python
ax = m.warp_functions[0].plot()
ax.legend(loc='best')
plt.title("RNA-seq",fontsize=20)
ax = m.warp_functions[1].plot()
ax.legend(loc='best')
plt.title("Bisulfite Seq",fontsize=20)
ax = m.warp_functions[2].plot()
ax.legend(loc='best')
plt.title("ATAC-seq",fontsize=20)
ax = m.warp_functions[3].plot()
ax.legend(loc='best')
plt.title("ChIP-seq",fontsize=20)
```

```python
LMR_inds = np.genfromtxt("p300_H3K27ac_inds.csv",delimiter=',',dtype=int)
cor_mat = m.correlation(0,2,LMR_inds[:,1],LMR_inds[:,0])
LMR_inds2 = np.genfromtxt("p300_H3K27ac_inds2.csv",delimiter=',',dtype=int)
cor_mat2 = m.correlation(1,2,LMR_inds2[:,1],LMR_inds2[:,0])
np.savetxt("p300_H3K27ac_inferred_cor.csv",np.diag(cor_mat),delimiter=",")
np.savetxt("p300_H3K27ac_inferred_cor2.csv",np.diag(cor_mat2),delimiter=",")
```

```python
sig_genes = m.find_markers([1])
sig_TFs = m.find_markers([3])
sig_TFs = np.array(sig_TFs)
sig_TFs = np.where(sig_TFs <= 0.05)[1]
sig_genes = np.array(sig_genes)
sig_genes = np.where(sig_genes <= 0.05)[1]
gene_TF_corr = m.correlation(1,3,sig_genes,sig_TFs)
np.savetxt("sig_TFs.csv",sig_TFs+1,delimiter=",")
np.savetxt("sig_genes.csv",sig_genes+1,delimiter=",")
np.savetxt("RNA_ATAC_inferred_corr.csv",gene_TF_corr,delimiter=",")
```

```python
atac_chip_corr = m.correlation(1,2,np.array([191,88,89,170,154,87,119,186,84,187,57,58]),np.array([15,16,17,23,24,25,55,58,60,59,56,57]),method="Spearman")
np.savetxt("ATAC_ChIP_inferred_corr.csv",atac_chip_corr,delimiter=",")
```

```python
corr_feat = m.infer_corresponding_features(0,2,np.array(range(169,175)),np.array(range(186,192)))
np.savetxt("rna_fit.csv",corr_feat[0],delimiter=",")
np.savetxt("atac_fit.csv",corr_feat[1],delimiter=",")
```
