```python
%matplotlib inline
%config InlineBackend.figure_format = 'png'
import matplotlib
import matplotlib.pyplot as plt
import GPy
import os
import numpy as np
import sklearn.preprocessing
from matcher import MATCHER
```

```python
X = np.genfromtxt("/proj/prins_lab/Josh/TCGA/Pan-Cancer/BRCA/TCGA_BRCA_705.csv",delimiter=',')
X = X.transpose()
num_rows = X.shape[0]
num_cols = X.shape[1]
X = X[1:num_rows,1:num_cols]
X = np.log(X+1)
#X = sklearn.preprocessing.scale(X)
```

```python
X2 = np.genfromtxt("/proj/prins_lab/Josh/TCGA/Pan-Cancer/BLCA/BLCA_expression.txt",delimiter='\t')
X2 = X2.transpose()
num_rows = X2.shape[0]
num_cols = X2.shape[1]
X2 = X2[1:num_rows,1:num_cols]
X2 = np.log(X2+1)
```

```python
X3 = np.genfromtxt("/proj/prins_lab/Josh/TCGA/Pan-Cancer/LUSC-LUAD/lung_expression.csv",delimiter=',')
X3 = X3.transpose()
num_rows = X3.shape[0]
num_cols = X3.shape[1]
X3 = X3[1:num_rows,1:num_cols]
X3 = np.log(X3+1)
```

```python
y = np.genfromtxt("/proj/prins_lab/Josh/TCGA/Pan-Cancer/BRCA/BRCA_dists.csv",delimiter=',')
y = y.reshape(-1,1)
```

```python
#gpr = GPy.models.GPRegression(y,X,GPy.kern.RBF(1,ARD=False))
gplvm = GPy.models.BayesianGPLVM(X3,input_dim=2,kernel=GPy.kern.RBF(2,ARD=False),num_inducing=10)
```

```python
gplvm.optimize(messages=True,max_iters=5000)
```

```python
m = MATCHER([X,X2,X3])
m.infer()
```

```python
ind = 6593
fig = m.model[0].plot_f(which_data_ycols=[ind],plot_data=True,plot_inducing=True)
fig.scatter(m.model[0].latent_space.mean,X[:,ind])
fig2 = m.model[1].plot_f(which_data_ycols=[ind],plot_data=True,plot_inducing=True)
fig2.scatter(m.model[1].latent_space.mean,X2[:,ind])
ind2 = 36
fig3 = m.model[2].plot_f(which_data_ycols=[ind2],plot_data=True,plot_inducing=True)
fig3.scatter(m.model[2].latent_space.mean,X3[:,ind2])
```
